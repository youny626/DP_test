```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/Users/zhiruzhu/Desktop/DP_libraries/DP_test/PUMS.csv")
# df = df.head(100)
df
```

```python
average_age = np.average(df["age"])
average_age
```

```python
import diffprivlib as dp

res = []
for eps in [0.01, 0.1, 1]:
    sensitivity = (df["age"].max() - df["age"].min()) / len(df)
    laplace = dp.mechanisms.Laplace(epsilon=eps, sensitivity=sensitivity)

    changes = []
    for i in range(len(df)):
        cur_df = df.copy()
        cur_df = cur_df.drop([i])
        assert len(cur_df) == len(df) - 1

        dp_average = laplace.randomise(np.average(cur_df["age"]))

        change = abs(average_age - dp_average)
        changes.append(change)

    res.append(changes)
```

```python
# eps = 0.01
pd.DataFrame(res[0]).describe()
```

```python
# eps = 0.1
pd.DataFrame(res[1]).describe()
```

```python
# eps = 1
pd.DataFrame(res[2]).describe()
```
