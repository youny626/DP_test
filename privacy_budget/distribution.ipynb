```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/Users/zhiruzhu/Desktop/DP_libraries/DP_test/PUMS.csv")
# df = df.head(100)
df
```

```python
average_age = np.average(df["age"])
average_age
```

```python
# computing how much the query result would change by removing each individual(record) out

changes = []
for i in range(len(df)):
    cur_df = df.copy()
    cur_df = cur_df.drop([i])
    assert len(cur_df) == len(df) - 1

    change = average_age - np.average(cur_df["age"])
    changes.append(change)
changes
```

```python
sorted_index = np.argsort(changes)
sorted_index
```

```python
sorted_changes = np.sort(changes)
sorted_changes
```

```python
np.quantile(sorted_changes, 0.10)
```

```python
idx1 = list(sorted_changes).index(np.quantile(sorted_changes, 0.25))
idx1
```

```python
idx2 = list(sorted_changes).index(np.quantile(sorted_changes, 0.75))
idx2
```

```python
sample1 = sorted_changes[:idx1]
sample2 = sorted_changes[idx2:]
```

```python
from scipy import stats

# two-sample Kolmogorov-Smirnov test
res = stats.ks_2samp(sample1, sample2)

res
# null hypothesis: the two samples come from the same distribution
# p-value = 0, we reject the null
```

```python
sample_index1 = sorted_index[:idx1]
sample_index2 = sorted_index[idx2:]
```

```python
# add DP noise
# try different epsilon values
# do ks test on the same two samples

import diffprivlib as dp
import tqdm

eps_list = np.arange(0.01, 1, 0.01, dtype=float) #[0.01, 0.1, 1, 10, 100]
sensitivity = (df["age"].max() - df["age"].min()) / len(df)

ks_test_results = []

for eps in tqdm.tqdm(eps_list):

    laplace = dp.mechanisms.Laplace(epsilon=eps, sensitivity=sensitivity)
    dp_average = laplace.randomise(np.average(average_age))

    changes = []

    for i in range(len(df)):
        cur_df = df.copy()
        cur_df = cur_df.drop([i])

        cur_dp_average = laplace.randomise(np.average(cur_df["age"]))

        change = dp_average - cur_dp_average
        changes.append(change)

    cur_sample1 = [changes[idx] for idx in sample_index1]
    cur_sample2 = [changes[idx] for idx in sample_index2]

    cur_res = stats.ks_2samp(cur_sample1, cur_sample2)
    ks_test_results.append(cur_res)

ks_test_results
```

```python
p_values = [res[1] for res in ks_test_results]
p_values
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8), dpi=200)
plt.plot(eps_list, p_values)
plt.xlabel("epsilon")
plt.ylabel("p_value")
plt.savefig("res")
plt.show()
```

```python
def fdr(p_values, q):
    # Based on https://matthew-brett.github.io/teaching/fdr.html

    # q = 0.1 # proportion of false positives we will accept
    N = len(p_values)
    sorted_p_values = np.sort(p_values) # sort p-values ascended
    i = np.arange(1, N+1) # the 1-based i index of the p values, as in p(i)
    below = (sorted_p_values < (q * i / N)) # True where p(i)<q*i/N
    if len(np.where(below)[0]) == 0:
        return False
    max_below = np.max(np.where(below)[0]) # max index where p(i)<q*i/N
    # print('p*:', sorted_p_values[max_below])

    # num_discoveries = 0
    for p in p_values:
        if p < sorted_p_values[max_below]:
            # print("Discovery: " + str(p))
            return True
            # num_discoveries += 1
    # print("num_discoveries =", num_discoveries)
    return False
```

```python
# add DP noise
# try different epsilon values
# do ks test on the same two samples

import diffprivlib as dp
import tqdm

eps_list = np.arange(0.01, 1, 0.01, dtype=float) #[0.01, 0.1, 1, 10, 100]
sensitivity = (df["age"].max() - df["age"].min()) / len(df)

# alpha = 0.05

q_list = np.arange(0.01, 0.1, 0.01, dtype=float)

eps_res = []
for q in q_list:

    best_eps = None

    for eps in tqdm.tqdm(eps_list[::-1]): # start from the largest eps

        laplace = dp.mechanisms.Laplace(epsilon=eps, sensitivity=sensitivity)

        rejects_null = False
        p_values = []

        for j in range(20):

            if rejects_null:
                # this eps rejects the null, so skip
                continue

            dp_average = laplace.randomise(np.average(average_age))

            changes = []

            for i in range(len(df)):
                # cur_df = df.copy()
                # cur_df = cur_df.drop([i])
                cur_df = dfs[i]

                cur_dp_average = laplace.randomise(np.average(cur_df["age"]))

                change = dp_average - cur_dp_average
                changes.append(change)

            cur_sample1 = [changes[idx] for idx in sample_index1]
            cur_sample2 = [changes[idx] for idx in sample_index2]

            cur_res = stats.ks_2samp(cur_sample1, cur_sample2)
            p_value = cur_res[1]
            p_values.append(p_value)

            # if p_value > alpha:
            #     # failed to reject null (two sample come from the same distribution)
            #     rejects_null = False
            # else:
            #     rejects_null = True

        # apply fdr for multiple comparisons
        # q = 0.05 # proportion of false positives we will accept
        if not fdr(p_values, q):
            # no discovery, so failed to reject null
            # print(eps)
            # print(p_values)
            best_eps = eps
            break

    eps_res.append(best_eps)

    # if rejects_null == False:
    #     # failed to reject null for all runs, this is the eps we want
    #     print(eps)
    #     print(p_values)
    #     break

print(eps_res)
```
