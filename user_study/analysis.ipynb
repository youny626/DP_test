```python
from scipy import stats

stats.ks_2samp(df["control1"].dropna(), df["test1"].dropna())
```

```python
stats.ks_2samp(df["control2"].dropna(), df["test2"].dropna())
```

```python
# The Mann-Whitney U test is a nonparametric test of the null hypothesis that the distribution underlying sample x is the same as the distribution underlying sample y. It is often used as a test of difference in location between distributions.

stats.mannwhitneyu(df["control1"].dropna(), df["test1"].dropna())
```

```python
stats.mannwhitneyu(df["control2"].dropna(), df["test2"].dropna())
```

```python
# The Kruskal-Wallis H-test tests the null hypothesis that the population median of all of the groups are equal. It is a non-parametric version of ANOVA.

stats.kruskal(df["control1"].dropna(), df["test1"].dropna())
```

```python
stats.kruskal(df["control2"].dropna(), df["test2"].dropna())
```

```python
def tests(sample1, sample2):
    print(stats.mannwhitneyu(sample1, sample2))
    print(stats.kruskal(sample1, sample2))
```

```python
def plot_hist(control, test, name):

    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.suptitle(name)

    ax1.hist(control)
    ax2.hist(test)

    ax1.set_title("control")
    ax2.set_title("test")

    ax1.set_xticks(np.arange(0.5, 5.1, 0.5))
    ax2.set_xticks(np.arange(0.5, 5.1, 0.5))
    y_max = max(control.value_counts().max(), test.value_counts().max()) + 0.1
    ax1.set_ylim(0, y_max)
    ax2.set_ylim(0, y_max)


    ax1.set(xlabel='eps', ylabel='count')
    ax2.set(xlabel='eps', ylabel='count')

    # Hide x labels and tick labels for top plots and y ticks for right plots.
    # ax1.label_outer()
    # ax2.label_outer()

    fig.tight_layout()

    # plt.savefig("hist1")
    plt.show()
    plt.close()
```

```python
from IPython.core.display_functions import display

# I don't know anything about differential privacy 23

c1 = df[df.initial_q == "I don't know anything about differential privacy"]["control1"].dropna()
t1 = df[df.initial_q == "I don't know anything about differential privacy"]["test1"].dropna()

# display(c1)
# display(t1)

tests(c1, t1)

plot_hist(c1, t1, "Q1 - I don't know anything about differential privacy")
```
