```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("DP_study_result_new_nov_18.csv")
df.columns = ["control1", "control2", "test1", "test2", "test_follow_up", "q3", "q4", "initial_q"]
df
```

```python
data = []
control1 = df["control1"].dropna()
control2 = df["control2"].dropna()
test1 = df["test1"].dropna()
test2 = df["test2"].dropna()

for eps1, eps2 in zip(control1, control2):
    data.append(["Control", eps1, eps2])

for eps1, eps2 in zip(test1, test2):
    data.append(["Treatment", eps1, eps2])

df1 = pd.DataFrame(data=data, columns=["group", "Q1", "Q2"])
df1
```

```python
from scipy import stats

stats.mannwhitneyu(control1, test1)
```

```python
stats.mannwhitneyu(control2, test2)
```

```python
df1 = df1.melt(id_vars="group", var_name="query", value_name="epsilon")
df1
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# plt.rcParams['figure.dpi'] = 300
sns.set_theme(style="ticks", palette="colorblind", font="Times New Roman", font_scale=1.6)
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"
fig_width, fig_height = plt.gcf().get_size_inches()
```

```python
# fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(35, 7))
from matplotlib.pyplot import figure

plt.rcParams["figure.figsize"] = [8, 3]

g = sns.swarmplot(data=df1, x="query", y="epsilon", hue="group", dodge=True, size=7, palette="tab10")
_ = g.set(xlabel=None, ylabel="Epsilon")
_ = g.set_yticks([i for i in np.arange(0.5, 5.1, 0.5)])
g.legend(fontsize='x-small', title_fontsize='x-small', loc='upper left', bbox_to_anchor=(0.35, 1))
g.legend_.set_title(None)
plt.tight_layout()
plt.savefig("figures/swarmplot.pdf", bbox_inches = 'tight', pad_inches = 0.05)
plt.show()
plt.close()
```

```python
df2 = df[df["initial_q"] == "I don't know anything about differential privacy"]
df2
```

```python
data = []

df2 = df[df["initial_q"] == "I don't know anything about differential privacy"]
control1 = df2["control1"].dropna()
control2 = df2["control2"].dropna()
test1 = df2["test1"].dropna()
test2 = df2["test2"].dropna()

for eps1, eps2 in zip(control1, control2):
    data.append(["Does not know DP", "Control", eps1, eps2])
for eps1, eps2 in zip(test1, test2):
    data.append(["Does not know DP", "Treatment", eps1, eps2])

df2 = df[df["initial_q"] != "I don't know anything about differential privacy"]
control1 = df2["control1"].dropna()
control2 = df2["control2"].dropna()
test1 = df2["test1"].dropna()
test2 = df2["test2"].dropna()

for eps1, eps2 in zip(control1, control2):
    data.append(["Knows DP", "Control", eps1, eps2])
for eps1, eps2 in zip(test1, test2):
    data.append(["Knows DP", "Treatment", eps1, eps2])

df2 = pd.DataFrame(data=data, columns=["know_dp", "group", "Q1", "Q2"])
df2
```

```python
df2 = df2.melt(id_vars=["know_dp", "group"], var_name="query", value_name="epsilon")
df2
```

```python
# fig_width, fig_height = plt.gcf().get_size_inches()
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 3), sharey=True)

g = sns.swarmplot(data=df2[df2["know_dp"] == "Does not know DP"], x="query", y="epsilon", hue="group", dodge=True, size=7, palette="tab10", ax=ax1)
_ = g.set(xlabel="Does not know DP", ylabel="Epsilon")
_ = g.set_yticks([i for i in np.arange(0.5, 5.1, 0.5)])
ax1.legend(fontsize='xx-small', loc='upper left', bbox_to_anchor=(0.16, 1))
g.legend_.set_title(None)
# ax1.get_legend().remove()

g = sns.swarmplot(data=df2[df2["know_dp"] == "Knows DP"], x="query", y="epsilon", hue="group", dodge=True, size=7, palette="tab10", ax=ax2)
_ = g.set(xlabel="Knows DP", ylabel=None)
_ = g.set_yticks([i for i in np.arange(0.5, 5.1, 0.5)])
# ax2.legend(fontsize='xx-small', loc='upper left', bbox_to_anchor=(0.15, 1.1))
# g.legend_.set_title(None)
ax2.get_legend().remove()


ax1.label_outer()
ax2.label_outer()

plt.tight_layout()
plt.savefig("figures/swarmplot_stratify.pdf", bbox_inches = 'tight', pad_inches = 0.05)
plt.show()
plt.close()
```
